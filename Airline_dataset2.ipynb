What is the flight status throughout the year?

```
# Convert 'Departure Date' column to datetime objects
df['Departure Date'] = pd.to_datetime(df['Departure Date'])

# Now you can extract the month
df["month"] = df["Departure Date"].dt.month
```

```
df.head()
```

```
# Create a list
status_unique = df['Flight Status'].unique()

# Subplots
num_rows = len(status_unique)
fig, axs = plt.subplots(num_rows, 1, figsize=(9, 5 * num_rows), sharex=False,gridspec_kw={'hspace': 0.5})

# Labels
labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# To iterate through the status and create separate graphs
for i, status in enumerate(status_unique):
    f_status = df[df['Flight Status'] == status]
    f_month = f_status.groupby(f_status['month'])['Flight Status'].count()

    bars = axs[i].bar(f_month.index, f_month.values)
    axs[i].set_title(f'Flight {status} per Month')
    axs[i].set_xlabel('Month')
    axs[i].set_ylabel('Nº of Flight')

    # To set up X-axis labels to display the months.
    axs[i].set_xticks(range(1, 13))  # Set label positions
    axs[i].set_xticklabels(labels, rotation=45, ha='right')  # Set labels and rotation

    # Add value labels to the bars.
    for bar in bars:
        height = bar.get_height()
        axs[i].annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

plt.show()
```

```
# some dates were formatted with "/" and others with "-". Same formatting must apply for all dates
# Convert the 'Departure Date' column back to string type before applying .str.replace
df['Departure Date'] = df['Departure Date'].astype(str).str.replace('-', '/')

# departure date dtype to datetime
df['Departure Date'] = pd.to_datetime(df['Departure Date'])
# gender dtype to category
df['Gender'] = df['Gender'].astype('category')

# list of columns to drop
to_drop = ['Pilot Name', 'Arrival Airport', 'Airport Continent']
df = df.drop(columns=to_drop)
df.info()
```

```
#we don't really need year because the dataset only conatins 2022
df['Year'] = df['Departure Date'].dt.year
#month & day could still be useful for determining busy periods in the year
df['Month'] = df['Departure Date'].dt.month
df['Day'] = df['Departure Date'].dt.day

#we want to sort the dates to allow for visuals
df = df.sort_values('Departure Date')
```

```
# getting the total number of flights for each date
accidents_per_date = df.groupby('Departure Date').size()
accidents_per_date
```

```
# plotting flights over time to see if we can determine most/least busy periods
accidents_per_date.plot(color='orange')
plt.show()
```
